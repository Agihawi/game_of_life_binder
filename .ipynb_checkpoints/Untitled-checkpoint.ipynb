```python
#import the random module
import random

while 1 == 1:
    #request input of the base number of branches --> attempt to make it an integer, could code it as an error message to give the user more detailed feedback.
    bases = int(input('Enter the base number of branches:'))
    #perform checks to make sure it's an odd number between 5 and 101
    if (bases >4) and (bases < 102) and (bases % 2 == 1):
        print('%s is a valid number of bottom branches' %bases)
    else:
        raise ValueError('Invalid input, input must be an odd number between 5 and 101')
    #print a message to seperate the christmas tree from string values above
    print('Generating your very own christmas tree...\n\n')
    #declare *spaces* - automatically sets to float, so make it an int
    spaces = 0 #int(bases+1)/2 provides maximum number of spaces
    #start off with the number of asterix equal to maximum base length
    num_asterix = bases
    #declare a list containing each of the correct combinations of spaces and *'s for each row of the tree
    tree_list = []
    #produce tree (in reverse order, bottom to top)
    while num_asterix > 0:
        row_of_tree=((int(spaces)*' ') + (int(num_asterix)*'*') + (int(spaces)*' '))
        tree_list.append(row_of_tree)
        num_asterix -=2
        spaces +=1
    #form a nested list containing the levels of each of the decorated tree levels
    reformed_tree = []
    #randomly mutate some * in the tree - loop through each level of the tree, then each term in the level
    for level in tree_list:
        new_level = ''
        for term in level:
            #randomly generate an integer and mutate the string - then add 'o' to the new level list 
            if (term == '*') and (random.randint(1,5) == 1):
                new_level +='o'
            else:
                new_level +=(term)
        reformed_tree.append(new_level)
    for term in reversed(reformed_tree):
        print(term)
    #add a stick on to the bottom of the tree
    print((int(spaces-1)*' ') + '|')

```
